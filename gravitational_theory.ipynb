# Gravity

This notebook is a short introduction to how gravity changes with altitude.

### Necessity of Accounting for Gravity Changes
1. For most intents and purposes, students or hobbyists working on a rocket won't need to worry about the magnitude of gravity changing with altitude for rocket flights to 30k ft, and even more so for flights to 10k ft. 
2. For high-precision apogee targeting (airbrakes), it's not a bad idea to account for the change in gravity with altitude, as it'll be about 0.3% weaker at 30k ft above ground level, but given that for tropospheric flights, drag is a much bigger force than gravity, considering gravity change still won't account for more than a couple of meters of height difference in your simulations. 
3. For space launches, the change in gravity with altitude is mcuh more significant. For example, the acceleration due to gravity on the ISS is about 89% of that on the surface of the Earth.

With that said, this notebook is more for interest than for practical design considerations.

Below is a plot of how gravity changes with altitude at Spaceport America, New Mexico:

```python
import gravity
from locations import location_SAC
import constants as con
import numpy as np
import matplotlib.pyplot as plt

# Plot strength of gravity at ground level, 5k ft, 10k ft, 15k ft, 20k ft, 25k ft, 30k ft, 35k ft
altitudes = np.array([0, 5000, 10000, 15000, 20000, 25000, 30000, 35000]) * con.ft_to_m_conversion
gravity_strength = np.zeros(altitudes.size)
for i in range(altitudes.size):
    gravity_strength[i] = gravity.get_local_gravity(location_SAC.latitude, altitudes[i] + location_SAC.elevation)

# use 2 x axis, one for feet and one for meters, with the ft axis on top
fig, ax1 = plt.subplots()
ax2 = ax1.twiny()
ax1.plot(altitudes * con.m_to_ft_conversion, gravity_strength)
ax1.set_xlabel('Altitude (AGL, ft)')
ax1.set_ylabel('Gravity (m/s^2)')
ax2.set_xlabel('Altitude (AGL, m)')
ax2.plot(altitudes, gravity_strength)
ax1.set_xlim(left=0)
ax2.set_xlim(left=0)
plt.title('Gravity Strength vs Altitude at Spaceport America')
plt.show()
```
